```python
def male_female_chlid(passanger):
    age,sex=passanger
    if age<=16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person']=titanic_df[['Age','Sex']].apply(male_female_chlid,axis=1)
```

```python
titanic_df[titanic_df['person']=='child']
```

```python
sns.factorplot('Pclass',data=titanic_df,hue='person',kind='count')
```
